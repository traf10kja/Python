# f(x) =0.6x^3+5.5x^2+10x -5

1. Определить корни

2. Найти интервалы, на которых функция возрастает

3. Найти интервалы, на которых функция убывает

4. Построить график

5. Вычислить вершину

6. Определить промежутки, на котором f > 0

7. Определить промежутки, на котором f < 0

```python
from sympy import *
a = 0.6
b = 5.5
c = 10
d = -5

x = Symbol("x")
f_x = a* (x**3) + b * (x**2) + c * (x**1) + d
f_x
```

```python
print("корень №1:", solve(f_x, x)[0].n(3))
print("корень №2:", solve(f_x, x)[1].n(3))
print("корень №3:", solve(f_x, x)[2].n(3))
```

```python
from sympy.plotting import plot

g1 = plot(f_x,(x,-8,2), line_color='red', title='График функции')
```

```python
derivate = diff(f_x)
derivate
```

```python
g2 = plot(derivate,(x,-8,2), line_color='green', title='График производной')
```

```python
g3 = plot(derivate, f_x,(x,-8,2), legend = True, xlabel = 'X', ylabel = 'Y', title='Общий график', show = False)

g3[0].line_color = 'green'
g3[1].line_color = 'red'

g3.show()
```

```python
print("Maximum X:", solve(derivate, x)[0].n(3))
print("Minimum X:", solve(derivate, x)[1].n(3))
```

```python
print("Maximum Y:", f_x.subs(x,-5))
print("Minimum Y:", f_x.subs(x,-1.11))

```

```python
solve(f_x > 0,x)
```

```python
solve(f_x < 0,x)
```

```python
solve(derivate < 0,x)
```

```python
solve(derivate > 0,x)
```
